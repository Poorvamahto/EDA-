```python
# Reading the other file
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# Now we will be combining the two tables based on the common column ="Country Code" using pandas
final_df= pd.merge(df, df_country, on='Country Code',how='left')
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
# to check the datatypes
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# making a pie chart of top 3 countries
plt.pie(country_val[:3],labels=country_names[:3],autopct= '%1.2f%%')
```

```python
#Observation = Zomato has the largest market base in India, followed by US and UK.

```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
# Observations= 
# Rating (4.5- 4.9) - Excellent
# Rating (4- 4.4)   - Ver Good
# Rating (3.5- 3.9) - Good
# Rating (2.5- 3.4) - Average
#-Rating (1.8-2.4)  - Poor
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating count", data= ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating count",hue='Rating color', data= ratings,palette = ['blue','red','orange','yellow','green','green'])
```
